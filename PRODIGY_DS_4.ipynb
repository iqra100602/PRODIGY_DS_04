```python
col_names1 = ['ID', 'Target', 'Sentiment', 'Statement']
df = pd.read_csv('twitter_training.csv', names=col_names1)
```

```python
df.dropna(axis=0 , inplace=True)
```

```python
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.shape
```

```python
sentiment_count = df['Sentiment'].value_counts()
sentiment_count
```

```python
plt.figure(figsize=(6, 3))
sentiment_count.plot(kind='bar', color=['purple', 'blue', 'red', 'black'])
plt.title('Distribution of sentiment')
plt.xlabel('Sentiment')
plt.ylabel('Total no. of Tweets')
plt.xticks(rotation='vertical')
plt.show()
```

```python
brand = df[df['Target'].str.contains('Google', case=False)]
brand_sentiment_count = brand['Sentiment'].value_counts()
brand_sentiment_count
```

```python
plt.figure(figsize=(6, 6))
plt.pie(brand_sentiment_count, labels=brand_sentiment_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Google Sentiment Distribution')
plt.show()
```
